```python
import random


names = [
        "James", "Miranda", 
        "Amina", "Parker",
        "Caleb", "Alvarado",
        "Blake", "Nash",
        "Chandler", "Walls",
        "Lilianna", "Gordon",
        "Karter", "Buckley",
        "Theodora", "Cervantes"
        "Kamari", "Daniels"
        "Ember", "Bean"
        "Mccoy", "Olson"
        "Isabel", "Hoffman"
        ]


def randomISBN():
    return f"{random.randint(10, 99)}-{random.randint(100000, 999999)}-{random.randint(0,9)}-{random.randint(0,9)}"


def randomYear():
    return str(random.randint(-5000, 0))[1:]+" B.C." if random.randint(-5000, 2023) < 0 else str(random.randint(0, 2023))+" A.D."


def randomName(data_names):
    rnd_one = random.randint(0, len(data_names)-1)
    rnd_two = random.randint(0, len(data_names)-1)

    while rnd_one == rnd_two:
        rnd_two = random.randint(0, len(data_names)-1)

    return f"{data_names[rnd_one]} {data_names[rnd_two]}"


def isBookGood(bool_data):
    if bool_data:
        return "Yes of course, it's a good book! ^_^"
    return "No, it's a bad book. Don't waste your time on this one !!!"


def randomList(generate=False):
    global rnd_list
    if generate:
        rnd_list = random.sample([
            ". Finance in general terms",
            ". Local politics, geopolitics and influence on the stock market",
            ". Investments in shares and real estate",
            ". Mutual funds, life and medical insurance",
            ". Technologies, innovations and potential risk",
            ". Acceptance of reality and ridiculous conditions",
            ". Slaves of modern society",
            ". Absence of other civilizations",
            ". The living filter"
            ], random.randint(5,9))


def iPrintList(data):
    for elem in data:
        print(elem)


def bookPresentation(book):
    print("="*70)
    print("""The book {}\nwas writend by {} in year {}\nand it's avalible for the price of {} $.\nMore info:\nISBN: {},\nNumber of pages: {},\nWorth reading: {}.""".format(
        book['Title'],
        book['Author'],
        book['Release year'],
        book['Price'],
        book['ISBN'],
        book['Number of pages'],
        isBookGood(book["Has value?"])
        ))

    print("\nBook content:", end='\n\n')
    iPrintList(book["Book content"])
    print("="*70, end=5*"\n")


book_one = {
    "Title": "Hello World - 5 Steps to Remote Work",
    "Author": "Igor Martin",
    "ISBN": "80-902734-1-6",
    "Price": 111.18,
    "Release year": 1939,
    "Number of pages": 512,
    "Has value?": True,
    "Book content": [
        "1. Finance in general terms",
        "2. Local politics, geopolitics and influence on the stock market",
        "3. Investments in shares and real estate",
        "4. Mutual funds, life and medical insurance",
        "5. Technologies, innovations and potential risk",
        "6. Acceptance of reality and ridiculous conditions",
        "7. Slaves of modern society",
        "8. Absence of other civilizations",
        "9. The living filter"
        ]
    }

book_one["Title"] = "Compound Interest Magic Science"
book_one["Author"] = "Alexander Markov"

# This generates a random book dict
def randomBook():
    return {
        "Title": book_one["Book content"][random.randint(0,8)][3:],
        "Author": randomName(names),
        "ISBN": randomISBN(),
        "Price": round(random.uniform(5,591), 2),
        "Release year": randomYear(),
        "Number of pages": random.randint(20, 1000),
        "Has value?": bool(random.getrandbits(1)),
        "none": randomList(True),
        "Book content": [str(rnd_list.index(elem)+1)+elem for elem in rnd_list]
        }


books_list = []

bookPresentation(book_one)
for _ in range(int(input("How many books do you want?: "))):
    books_list.append(bookPresentation(randomBook()))

```
